```python
import matplotlib.pyplot as plt
plt.scatter(perch_length, perch_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
# perch_weight가 target data
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    perch_length, perch_weight, random_state=42)
# 회귀문제라 stratify 매개변수를 사용 않함. 임의 숫자가 target이라서 일반적으로 사용 않함.

print(train_input.shape, test_input.shape)
```

```python
# sklearn은 배열이 2차원일거 기대를 함. 그래서 reshape 해줘야함
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)
```

```python
print(train_input.shape, test_input.shape)
```

결정 계수 ($ R^2$)

image.png

```python
# 만약 타겟의 평균 정도를 예측하는 수준이라면
# 즉 분자와 분모가 비슷해져 R Square는 0에 가까워진다)
# 예측이 타겟에 아주 가까워지면(분자가 0에 가까워지기 떄문, 1에 가까운 값이됨)
# R Square 는 1에 가까울수록 좋은 모델이라고함..
```

```python
from sklearn.neighbors import KNeighborsRegressor
knr = KNeighborsRegressor()
# k-최근접 이웃 회귀 모델을 훈련합니다
knr.fit(train_input, train_target)
```

```python
# 테스트 데이터를 넣어서 점수 확인
knr.score(test_input, test_target)
```

```python
from sklearn.metrics import mean_absolute_error
# 테스트 세트에 대한 예측을 만듭니다
test_prediction = knr.predict(test_input)
# 테스트 세트에 대한 평균 절댓값 오차를 계산합니다
mae = mean_absolute_error(test_target, test_prediction)
print(mae)
```

```python
# 과대적합 vs 과소적합
print(knr.score(train_input, train_target))
knr.score(test_input, test_target)

# 일반적으로는 훈련점수가 높고 테스트 점수가 낮게 나와야 합니다.

# 위 같은 모양을 훈련 테스트가 과소적합되었다고 합니다 uderfiting
# 예) 훈련 0.99 테스트 0..6 이면 과대적합이라고 합니다 overfiting
```

```python
# 이웃의 갯수를 3으로 설정합니다
knr.n_neighbors = 3
# 모델을 다시 훈련합니다
knr.fit(train_input, train_target)
print(knr.score(train_input, train_target))
# *이웃의 갯수를 줄이면 과대 적합이 되고, 갯수를 늘리면 과대 적합이된다
```

```python
print(knr.score(test_input, test_target))
```

확인문제

```python
# k-최근접 이웃 회귀 객체를 만듭니다
knr = KNeighborsRegressor()
# 5에서 45까지 x 좌표를 만듭니다
x = np.arange(5, 45).reshape(-1, 1)

# n = 1, 5, 10일 때 예측 결과를 그래프로 그립니다.
for n in [1, 5, 42]:
    # 모델 훈련
    knr.n_neighbors = n
    knr.fit(train_input, train_target)
    # 지정한 범위 x에 대한 예측 구하기 
    prediction = knr.predict(x)
    # 훈련 세트와 예측 결과 그래프 그리기
    plt.scatter(train_input, train_target)
    plt.plot(x, prediction)
    plt.title('n_neighbors = {}'.format(n))    
    plt.xlabel('length')
    plt.ylabel('weight')
    plt.show()
```
